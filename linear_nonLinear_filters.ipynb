```python
import cv2
import numpy as np
```

```python
img=cv2.imread('/content/baby.jpeg')
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
#gussian filter
#coverting the image to gray:
g_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(g_img)
gussian_filter_img=cv2.GaussianBlur(g_img,(5,5),0)
cv2_imshow(gussian_filter_img)
```

```python
#mean filter
kernal=np.ones((5,5),np.float32)/25
mean_filter=cv2.filter2D(g_img,-1,kernal)
cv2_imshow(mean_filter)
```

```python
#non linear filter:
#bilteral
#minimum
#maximum
# Median
```

```python
#medainBlurr:
Medain_blurr=cv2.medianBlur(g_img,5)
cv2_imshow(Medain_blurr)
```

```python
#bilatral filter:
b_filter=cv2.bilateralFilter(g_img,15,75,75)
cv2_imshow(b_filter)
```

```python
# minimum filter
im2=cv2.imread('/content/512x512bb.jpg')
cv2_imshow(im2)
min_filter=cv2.erode(im2,cv2.getStructuringElement(cv2.MORPH_RECT,(5,5)))
cv2_imshow(min_filter)
```

```python
max_filter=cv2.dilate(im2,cv2.getStructuringElement(cv2.MORPH_RECT,(5,5)))
cv2_imshow(max_filter)
```
