```python
import cv2
```

```python
img1=cv2.imread('/content/images (1).jpeg')
```

```python
img2=cv2.imread('/content/images (2).jpeg')
```

```python
cv2_imshow(img1)
```

```python
gray_img=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
blur_img=cv2.GaussianBlur(gray_img,(5,5),0)
cv2_imshow(blur_img)
```

```python
canny_edges1=cv2.Canny(blur_img,50,50)
canny_edges2=cv2.Canny(blur_img,50,150)
```
